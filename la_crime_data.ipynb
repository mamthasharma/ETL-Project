```python
import pandas as pd
import numpy as np
import requests
from pprint import pprint
```

```python
file_to_load = 'LACityCrime.csv'
```

```python
# Read Crime File and store into Pandas data frame
crime_LA_df = pd.read_csv(file_to_load)
```

```python
crime= len(crime_LA_df["DR_NO"].value_counts())
crime_count=pd.DataFrame([crime], columns = ["Total Crimes Recorded"])
crime_count
```

```python
LA_crime_df= crime_LA_df[["DR_NO","Crm Cd","Crm Cd Desc","Date Rptd","Premis Desc"]].copy()
LA_crime_df.head()
```

```python
LA_crime_df['City']='LA'
```

```python
LA_crime_df.head()
```

```python
# df.loc[df['column_name'].isin(some_values)]
gta_df=LA_crime_df.loc[LA_crime_df["Crm Cd"].isin([510,487])]
```

```python
gta_df.head()
```

```python
gta_df['Date']=pd.to_datetime(gta_df['Date Rptd'].str.split().map(lambda x:x[0]))
```

```python
gta_df.head()
```

```python
gta_df['PrimaryDesc']='Grand Theft Auto'
```

```python
# df.rename(columns={"A": "a", "B": "c"})
gta_df=gta_df.rename(columns={"DR_NO":"ComplaintNum","Crm Cd Desc":"SecondaryDesc","Premis Desc":"Premises"})
gta_df.head()
```

```python
gta_df=gta_df[["ComplaintNum","PrimaryDesc","SecondaryDesc","Premises","Date","City"]].copy()
```
